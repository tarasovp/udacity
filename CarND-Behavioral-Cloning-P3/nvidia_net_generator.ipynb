```python
import pandas as pd
import cv2
import numpy as np

import os

os.environ['CUDA_VISIBLE_DEVICES']='0'
```

```python
q=np.array([[0.299,0.587,0.114],\
           [-0.14713, -0.28886,0.436],\
           [0.612, -0.51499, -0.10001]])

def preproccess_image(image):
    return (np.dot(image,q.T)+[-128,0,0])/2
    

```

```python
datadirs=['/notebooks/udacity/car_data/data/',\
                 '/notebooks/udacity/car_data/udacity_from sides/',
                 '/notebooks/udacity/car_data/udacity_images/']
```

```python
dfs=[]

for folder in datadirs:
    df=pd.read_csv(folder+'driving_log.csv', skiprows=1,\
            names=['center','left','right','steering','trottle','brake','speed'])
    
    df['center']=df.center.apply(lambda x:folder+'IMG/'+x.split('/')[-1])
    df['left']=df.left.apply(lambda x:folder+'IMG/'+x.split('/')[-1])
    df['right']=df.right.apply(lambda x:folder+'IMG/'+x.split('/')[-1])
    dfs.append(df)
    
df=pd.concat(dfs,axis=0)
```

```python
images=[]
corr=0.25

for a in [('center',0),('left',corr),('right',-corr)]:
    images.extend(df[df.steering!=0][[a[0],'steering']].apply(lambda x:[x[0],x[1]+a[1],1],axis=1).values)
    images.extend(df[df.steering!=0][[a[0],'steering']].apply(lambda x:[x[0],-1*(x[1]+a[1]),-1],axis=1).values)
```

```python
import random

random.shuffle(images)

from sklearn.cross_validation import train_test_split

images_train, images_test = train_test_split(np.array(images), test_size=0.15)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.hist([a[1] for a in images],bins=100)
plt.show()
```

```python
batch_size=128

for a in range(0,1000,batch_size):
    if a+batch_size<1000:
        print (a,a+batch_size)
    
```

```python
import cv2
def generator(images, batch_size):
    while 1:
        for i in range(0,len(images),batch_size):
            if i+batch_size<len(images):
                ims=np.array([preproccess_image(cv2.imread(images[i+j][0])) \
                              for j in range(batch_size)])
                for j in range(batch_size):
                    if images[i+j][2]==-1:
                        ims[i+j]=ims[:, ::-1]
                yield (ims, np.array(images[i:i+batch_size][:,1]))
```

```python
from keras.models import Sequential
from keras.layers import Flatten,Dense, Convolution2D, Activation, MaxPooling2D, Lambda,Conv2D,Dropout, Cropping2D
```

```python
from keras.callbacks import History,TensorBoard
history = History()

tb=TensorBoard(log_dir='./logs', histogram_freq=0, write_graph=True, write_images=False)
```

```python

model=Sequential()

model.add(Cropping2D(cropping=((50,20), (0,0)), input_shape=(160,320,3)))
#model.add(Lambda(lambda x: (x / 255.0) - 0.5))


model.add(Conv2D(24, kernel_size=(5, 5),
                 activation='relu',
                  strides =(2,2),\
                 padding='Same'))

model.add(Dropout(0.5))

model.add(Conv2D(36, kernel_size=(5, 5),
                 activation='relu',
                  strides =(2,2),\
                 padding='Same'))
model.add(Dropout(0.5))


model.add(Conv2D(48, kernel_size=(5, 5),
                 activation='relu',
                  strides =(2,2),\
                 padding='Same'))
model.add(Dropout(0.5))

model.add(Conv2D(64, kernel_size=(3, 3),
                 activation='relu',
                  strides =1,\
                 padding='Same'))
#model.add(Dropout(0.5))

model.add(Conv2D(64, kernel_size=(3, 3),
                 activation='relu',
                  strides =1,\
                 padding='Same'))
#model.add(Dropout(0.5))

model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                  strides =1,\
                 padding='Same'))
model.add(Dropout(0.5))

model.add(Flatten())


model.add(Dense(1164, activation='relu'))
#model.add(Dropout(0.5))

model.add(Dense(100, activation='relu'))

model.add(Dropout(0.5))

model.add(Dense(50, activation='relu'))

model.add(Dense(10, activation='relu'))
#model.add(Dropout(0.5))

model.add(Dense(1))
```

```python
model.compile(loss='mse',optimizer='adam')
```

```python
train_gen=generator(images_train, 256)
test_gen=generator(images_test,256)
```

```python
model.fit_generator(train_gen, steps_per_epoch= len(images_train)/256, \
                    validation_data=test_gen, \
            validation_steps=len(images_test)/256, epochs=100,callbacks=[history, tb])

#model.fit_generator(generator.flow(X[:1000], y[:1000]), validation_data=(X_test, y_test),\
#        steps_per_epoch=1000,epochs=100,callbacks=[history, tb])
```
